```python
df1=pd.read_csv('Social_Network_Ads.csv')
print(df1)
```

```python
# Creat X and Y

x=df1.iloc[:,2:4].values
x
```

```python
y=df1.iloc[:,4].values
y
```

```python
# Feature Scaling - Standardization (To Standardize Salary)

from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
print(x)
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Creat Model KNN with N-Neighbors=5

from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
KNN.fit(x_train,y_train)
print(KNN)
```

```python
# prediction for x

y_pred=KNN.predict(x)
print(y_pred)
```

```python
# Display Accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(y,y_pred))
```

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y,y_pred)
print(cm)
```

```python
# Display ROC - AUC Curve

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,threshold=roc_curve(y,y_pred)
roc_df=pd.DataFrame(zip(fpr,tpr,threshold),columns=['FPR','TPR','Thrshold'])
display(roc_df)
```

```python
# Plot ROC AOC Graph

plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=1)
KNN.fit(x_train,y_train)
y_pred=KNN.predict(x)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,threshold=roc_curve(y,y_pred)
roc_df=pd.DataFrame(zip(fpr,tpr,threshold),columns=['FPR','TPR','THRESHOLD'])
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print('Accuracy')
print(accuracy_score(y,y_pred))
print('Confusion Matrix')
cm=confusion_matrix(y,y_pred)
print(cm)
print(roc_df)
```

```python
# Plot ROC AUC Graph

plt.plot(fpr,tpr,c='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
# Creat Model Logistics Regression

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,threshold=roc_curve(y,y_pred)
roc_df=pd.DataFrame(zip(fpr,tpr,threshold),columns=['FPR','TPR','Threshold'])
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print('Accuracy')
print(accuracy_score(y,y_pred))
print('Confusion Matrix')
cm=confusion_matrix(y,y_pred)
print(cm)
display(roc_df)
```
